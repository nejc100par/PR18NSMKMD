# Klasifikcaija

```python
import numpy as np
import pandas as pd
import Orange
from sklearn.metrics import accuracy_score
```

```python
data = Orange.data.Table("data/novi_pod/pn13-17_full.csv")
```

```python
train = data[:int(0.7*len(data))]
test = data[int(0.7*len(data)):]
```

## Večinski

```python
majorityLearner = Orange.classification.majority.MajorityLearner()
asd = majorityLearner(train)
asd
```

# SklTree Learner

```python
learner = Orange.classification.tree.TreeLearner()
```

```python
classi = learner(train)
```

```python
data.domain.class_var
print("Accuracy:", accuracy_score(test.Y, classi(test)))
```

# Povezovalna

```python
learnerlearner = Orange.classification.CN2Learner()
learnerlearner.rule_finder.search_algorithm.beam_width = 10
learnerlearner.rule_finder.search_strategy.bound_continuous = True
learnerlearner.rule_finder.general_validator.min_covered_examples = 15
learnerlearner.rule_finder.general_validator.max_rule_length = 3
classifier = learnerlearner(data)
```

```python
for rule in classifier.rule_list:
    print(rule, rule.curr_class_dist.tolist())
```
